```
import requests
from bs4 import BeautifulSoup as soup 
import pandas as pd

session = requests.Session()
links = []
articles = []

url = 'https://www.greenbiz.com/topics/renewable-energy?page={}'
for page_num in range(0, 300):
    page_url = url.format(page_num)
    r = session.get(page_url)
    b = soup(r.content, 'lxml')
    
    articles_on_page = b.findAll('h2', {'class': 'article-teaser-horizontal__title'})
    
    if not articles_on_page:
        # no more articles on this page or all pages have been scraped
        break
    
    for news in articles_on_page:
        link = news.a['href']
        if not any(x in link for x in ['/whitepaper/', '/webcast/', '/video/', '/report/']):
            links.append(link)
            articles.append({'heading': news.text, 'body': ''})
    

for i, link in enumerate(links):
    page = session.get('https://www.greenbiz.com'+link)
    bsobj = soup(page.content)
    article_body = bsobj.find('div', {'class': 'article__body'}) or bsobj.find('div', {'class': 'article7__body'})
    
    if article_body:
        articles[i]['body'] = article_body.text.strip()

df = pd.DataFrame(articles)

```
