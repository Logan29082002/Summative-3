```
import requests
from bs4 import BeautifulSoup
import pandas as pd

data = []
for page in range(1, 151):  # scrape the first three pages
    url = f"https://www.theguardian.com/environment/renewableenergy?page={page}"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    articles = soup.find_all("div", class_="fc-item__container")

    for article in articles:
        headline = article.find("span", class_="js-headline-text").text
        date = article.find("time", class_="fc-item__timestamp")["datetime"]
        author_element = article.find("span", class_="fc-item__byline__author-name")#.text
        author = author_element.text if author_element else "Unknown"
        content_url = article.find("a")["href"]#("a", class_="js-headline-text")
        
        # get full text of article
        content_response = requests.get(content_url)
        content_soup = BeautifulSoup(content_response.content, "html.parser")
        article_body_element = content_soup.find("div", class_="article-body-commercial-selector")
        article_body = article_body_element.text.strip() if article_body_element is not None else "No content found"
        #article_body = content_soup.find("div", class_="article-body-commercial-selector").text.strip()

        data.append({
            "headline": headline,
            "date": date,
            "author": author,
            "content_url": content_url,
            "article_body": article_body
        })

df = pd.DataFrame(data)

```
