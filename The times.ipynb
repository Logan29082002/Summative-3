```
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the URL to scrape
url = 'https://www.telegraph.co.uk/politics/'

# Make a request to the website
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find the section of the page that contains the article links
article_list = soup.find('section', {'class': 'list-of-entities__items'}).find_all('li')

# Define empty lists to store the article data
headlines = []
summaries = []
dates = []
urls = []

# Loop through each article and extract the relevant data
for article in article_list:
    
    # Extract the headline
    headline = article.find('a', {'class': 'list-of-entities__title-link'}).text
    headlines.append(headline)
    
    # Extract the summary
    summary = article.find('div', {'class': 'list-of-entities__summary'}).text
    summaries.append(summary)
    
    # Extract the date
    date = article.find('time')['datetime']
    dates.append(date)
    
    # Extract the URL
    url = article.find('a', {'class': 'list-of-entities__title-link'})['href']
    urls.append(url)

# Combine the data into a Pandas DataFrame
data = {'Headline': headlines, 'Summary': summaries, 'Date': dates, 'URL': urls}
df = pd.DataFrame(data)

# Print the DataFrame
print(df)
```
